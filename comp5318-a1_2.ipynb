## **1. Data Pre-processing**

```python
# Import all libraries
from sklearn.model_selection import StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
import numpy as np
```

```python
# Ignore future warnings
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
```

```python
# Load the rice dataset: rice-final2.csv
# df = pd.read_csv("test-before.csv", na_values='?')
df = pd.read_csv("rice-final2.csv", na_values='?')
df.head()
```

```python
# Pre-process dataset
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# 1. Filling in the missing attribute values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# 2. Normalizing the data
scaler = MinMaxScaler()
X = scaler.fit_transform(X_imputed)

# 3. Changing the class values: class1 -> 0, class2 -> 1
y = y.replace({'class1': 0, 'class2': 1}).astype(int)
y = y.to_numpy()
```

```python
# Print first ten rows of pre-processed dataset to 4 decimal places as per assignment spec
# A function is provided to assist

def print_data(X, y, n_rows=10):
    """Takes a numpy data array and target and prints the first ten rows.
    
    Arguments:
        X: numpy array of shape (n_examples, n_features)
        y: numpy array of shape (n_examples)
        n_rows: numpy of rows to print
    """
    for example_num in range(n_rows):
        for feature in X[example_num]:
            print("{:.4f}".format(feature), end=",")

        if example_num == n_rows - 1:
            print(y[example_num],end="")
        else:
            print(y[example_num])
            
```

```python
print_data(X, y)
```

## **2. Build Classifiers**

- Part 1:  Logistic Regression, Na茂ve Bayes
- Part 2:  KNN, Decision Tree, Ada Boost, Gradient Boost, Random Forest, SVM

### Part 1: Cross-validation without parameter tuning

```python
## Setting the 10 fold stratified cross-validation
cvKFold=StratifiedKFold(n_splits=10, shuffle=True, random_state=0)

# The stratified folds from cvKFold should be provided to the classifiers
```

```python
# Logistic Regression
log_reg = LogisticRegression(random_state=0, max_iter=1000)

logr_scores = cross_val_score(log_reg, X, y, cv=cvKFold, scoring='accuracy')
logr_acc = logr_scores.mean()
```

```python
# Na茂ve Bayes
```

```python
nb = GaussianNB()

nb_scores = cross_val_score(nb, X, y, cv=cvKFold, scoring='accuracy')
nb_acc = nb_scores.mean()
```

### Part 1 Results


```python
# Print results for each classifier in part 1 to 4 decimal places here:
print("LogR average cross-validation accuracy: {:.4f}".format(logr_acc))
print("NB average cross-validation accuracy: {:.4f}".format(nb_acc))
```

### Part 2: Cross-validation with parameter tuning

```python
# KNN 
# parameters you may consider
k = [1, 3, 5, 7]
p = [1, 2]

# split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=0
)

# define the KNN classifier
knn = KNeighborsClassifier()

# define parameter grid
knn_param_grid = {
    'n_neighbors': k,
    'p': p
}

# grid search with 10-fold stratified cross-validation
knn_grid = GridSearchCV(
    estimator=knn,
    param_grid=knn_param_grid,
    cv=cvKFold,
    scoring='accuracy'
)

# fit on training data
knn_grid.fit(X_train, y_train)

# best model and results
knn_best_params = knn_grid.best_params_
knn_best_cv_acc = knn_grid.best_score_

# test set prediction
y_pred_knn = knn_grid.predict(X_test)
knn_test_acc = accuracy_score(y_test, y_pred_knn)
```

```python
# Decision Tree 
# parameters you may consider
max_depth = [3, 5, 7, 10]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]

# split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=0
)

# define the decision tree classifier
dt = DecisionTreeClassifier(random_state=0)

# define parameter grid
dt_param_grid = {
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf
}

# grid search with 10-fold stratified cross-validation
dt_grid = GridSearchCV(
    estimator=dt,
    param_grid=dt_param_grid,
    cv=cvKFold,
    scoring='accuracy'
)

# fit on training data
dt_grid.fit(X_train, y_train)

# best model and results
dt_best_params = dt_grid.best_params_
dt_best_cv_acc = dt_grid.best_score_

# test set prediction
y_pred_dt = dt_grid.predict(X_test)
dt_test_acc = accuracy_score(y_test, y_pred_dt)
```

```python
from sklearn.ensemble import AdaBoostClassifier
# Ada Boost
# parameters you may consider
n_estimators = [50, 100, 150]
learning_rate = [0.1, 0.2, 0.3, 0.5]

# define the AdaBoost classifier
ada = AdaBoostClassifier(random_state=0)

# define parameter grid
ada_param_grid = {
    'n_estimators': n_estimators,
    'learning_rate': learning_rate
}

# grid search with 10-fold stratified cross-validation
ada_grid = GridSearchCV(
    estimator=ada,
    param_grid=ada_param_grid,
    cv=cvKFold,
    scoring='accuracy'
)

# fit on training data
ada_grid.fit(X_train, y_train)

# best model and results
ada_best_params = ada_grid.best_params_
ada_best_cv_acc = ada_grid.best_score_

# test set prediction
y_pred_ada = ada_grid.predict(X_test)
ada_test_acc = accuracy_score(y_test, y_pred_ada)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
# Gradient Boost
# parameters you may consider
max_depth = [1, 3, 5, 7]
n_estimators = [50, 100, 150]
learning_rate = [0.1, 0.2, 0.3, 0.5]

# define the gradient boosting classifier
gb = GradientBoostingClassifier(random_state=0)

# define parameter grid
gb_param_grid = {
    'max_depth': max_depth,
    'n_estimators': n_estimators,
    'learning_rate': learning_rate
}

# grid search with 10-fold stratified cross-validation
gb_grid = GridSearchCV(
    estimator=gb,
    param_grid=gb_param_grid,
    cv=cvKFold,
    scoring='accuracy'
)

# fit on training data
gb_grid.fit(X_train, y_train)

# best model and results
gb_best_params = gb_grid.best_params_
gb_best_cv_acc = gb_grid.best_score_

# test set prediction
y_pred_gb = gb_grid.predict(X_test)
gb_test_acc = accuracy_score(y_test, y_pred_gb)
```

```python
# Random Forest
# You should use RandomForestClassifier from sklearn.ensemble with information gain and max_features set to 'sqrt'
# parameters you may consider
n_estimators = [10, 30, 60, 100]
max_leaf_nodes = [6, 12]

# define the random forest classifier
rf = RandomForestClassifier(criterion='entropy', max_features='sqrt', random_state=0)

# define parameter grid
rf_param_grid = {
    'n_estimators': n_estimators,
    'max_leaf_nodes': max_leaf_nodes
}

# grid search with 10-fold stratified cross-validation
rf_grid = GridSearchCV(
    estimator=rf,
    param_grid=rf_param_grid,
    cv=cvKFold,
    scoring='accuracy'
)

# fit on training data
rf_grid.fit(X_train, y_train)

# best model and results
rf_best_params = rf_grid.best_params_
rf_best_cv_acc = rf_grid.best_score_

# test set prediction
y_pred_rf = rf_grid.predict(X_test)
rf_test_acc = accuracy_score(y_test, y_pred_rf)
rf_macro_f1 = f1_score(y_test, y_pred_rf, average='macro')
rf_weighted_f1 = f1_score(y_test, y_pred_rf, average='weighted')
```

```python
# SVM
# parameters you may consider
C = [0.01, 0.1, 1, 5]
gamma = [0.01, 0.1, 1, 10]
kernel = ['rbf']

# define the SVM classifier
svm = SVC()

# define parameter grid
svm_param_grid = {
    'C': C,
    'gamma': gamma,
    'kernel': kernel
}

# grid search with 10-fold stratified cross-validation
svm_grid = GridSearchCV(
    estimator=svm,
    param_grid=svm_param_grid,
    cv=cvKFold,
    scoring='accuracy'
)

# fit on training data
svm_grid.fit(X_train, y_train)

# best model and results
svm_best_params = svm_grid.best_params_
svm_best_cv_acc = svm_grid.best_score_

# test set prediction
y_pred_svm = svm_grid.predict(X_test)
svm_test_acc = accuracy_score(y_test, y_pred_svm)
```
