# Data Preprocessing

## merge CER

```python
# 1. 去掉列名空格
df_accredited.columns = df_accredited.columns.str.strip()
df_committed.columns = df_committed.columns.str.strip()
df_probable.columns = df_probable.columns.str.strip()

# 2. rename
df_accredited_full = df_accredited.rename(columns={
    "Power station name": "project_name",
    "State": "state",
    "Installed capacity (MW)": "capacity_mw",
    "Fuel Source (s)": "fuel_source",
    "Accreditation code": "accreditation_code",
    "Postcode": "postcode",
    "Accreditation start date": "accreditation_start_date",
    "Approval date": "approval_date"
})

df_committed_full = df_committed.rename(columns={
    "Project Name": "project_name",
    "State": "state",
    "MW Capacity": "capacity_mw",
    "Fuel Source": "fuel_source",
    "Committed Date (Month/Year)": "committed_date"
})

df_probable_full = df_probable.rename(columns={
    "Probable Projects": "project_name",
    "State": "state",
    "MW Capacity": "capacity_mw",
    "Fuel Source": "fuel_source"
})

# 3. concat（三个表一起）
df_cer = pd.concat(
    [df_accredited_full, df_committed_full, df_probable_full],
    ignore_index=True
)

# 4. 标记来源
df_cer.loc[df_cer["accreditation_code"].notna(), "project_status"] = "accredited"
df_cer.loc[df_cer["committed_date"].notna(), "project_status"] = "committed"

# probable（两列都为空）
df_cer.loc[
    df_cer["accreditation_code"].isna() & df_cer["committed_date"].isna(),
    "project_status"
] = "probable"

# 5. 查看结果
print(df_cer.head())
print(df_cer.info())
```

Clean NGER

```python
df_nger_clean = df_nger.copy()

# 先统一列名格式：去空格
df_nger_clean.columns = df_nger_clean.columns.str.strip()

# 统一映射到标准字段名
nger_rename_map = {
    "Reporting entity": "reporting_entity",
    "Facility name": "facility_name",
    "Type": "facility_type",
    "State": "state",
    "Electricity production GJ": "electricity_production_gj",
    "Electricity production MWh": "electricity_production_mwh",
    "Total scope 1 emissions t CO2 e": "scope1_emissions_tco2e",
    "Total scope 2 emissions t CO2 e": "scope2_emissions_tco2e",
    "Total emissions t CO2 e": "total_emissions_tco2e",
    "Emission intensity t CO2 e MWh": "emission_intensity_tco2e_per_mwh",
    "Grid connected": "grid_connected",
    "Grid Connected": "grid_connected",
    "Primary fuel": "primary_fuel",
    "Primary Fuel": "primary_fuel",
    "Important notes": "important_notes",
    "Important Notes": "important_notes",
    "Controlling corporation": "controlling_corporation",
    "Electricity Production Mwh": "electricity_production_mwh",
    "Emission Intensity t Mwh": "emission_intensity_tco2e_per_mwh",
    "Total Scope 2 Emissions t CO2 e 2": "scope2_emissions_tco2e",
    "Grid Connected2": "grid_connected",
    "Scope 1 t CO2 e": "scope1_emissions_tco2e",
    "Scope 2 t CO2 e": "scope2_emissions_tco2e",
}

df_nger_clean = df_nger_clean.rename(columns=nger_rename_map)

# 如果同一个标准字段名因为 rename 后重复出现，合并成一列：
# 同名列按从左到右用 combine_first 合并
def coalesce_duplicate_columns(df):
    df = df.copy()
    cols = pd.Series(df.columns)
    for col in cols[cols.duplicated()].unique():
        same_cols = df.loc[:, df.columns == col]
        merged = same_cols.bfill(axis=1).iloc[:, 0]
        df = df.loc[:, df.columns != col].copy()
        df[col] = merged
    return df

df_nger_clean = coalesce_duplicate_columns(df_nger_clean).copy()

# 只保留后续分析需要的字段
nger_keep_cols = [
    "year",
    "reporting_entity",
    "facility_name",
    "facility_type",
    "state",
    "electricity_production_gj",
    "electricity_production_mwh",
    "scope1_emissions_tco2e",
    "scope2_emissions_tco2e",
    "total_emissions_tco2e",
    "emission_intensity_tco2e_per_mwh",
    "grid_connected",
    "primary_fuel",
    "important_notes",
    "controlling_corporation"
]

existing_nger_cols = [c for c in nger_keep_cols if c in df_nger_clean.columns]
df_nger_clean = df_nger_clean[existing_nger_cols].copy()
#print("NGER cleaned shape:", df_nger_clean.shape)

# 文本清洗
nger_text_cols = [
    c for c in [
        "reporting_entity", "facility_name", "facility_type", "state",
        "grid_connected", "primary_fuel", "important_notes", "controlling_corporation"
    ] if c in df_nger_clean.columns
]

for col in nger_text_cols:
    df_nger_clean[col] = df_nger_clean[col].astype(str).str.strip()

# 无效值替换
df_nger_clean.replace({"-": pd.NA, "nan": pd.NA, "None": pd.NA}, inplace=True)

# 数值列转换
nger_numeric_cols = [
    c for c in [
        "year",
        "electricity_production_gj",
        "electricity_production_mwh",
        "scope1_emissions_tco2e",
        "scope2_emissions_tco2e",
        "total_emissions_tco2e",
        "emission_intensity_tco2e_per_mwh"
    ] if c in df_nger_clean.columns
]

for col in nger_numeric_cols:
    df_nger_clean[col] = pd.to_numeric(df_nger_clean[col], errors="coerce")

# 统一州字段，只保留有效州
valid_states = ["NSW", "VIC", "QLD", "SA", "WA", "TAS", "NT", "ACT"]
df_nger_clean["state"] = df_nger_clean["state"].astype(str).str.upper().str.strip()
df_nger_clean.loc[~df_nger_clean["state"].isin(valid_states), "state"] = pd.NA

# 去掉没有州的记录（因为后面要按州 join）
df_nger_clean = df_nger_clean.dropna(subset=["state"]).copy()

print("=== df_nger_clean info ===")
print(df_nger_clean.info())
df_nger_clean.head()
```

Clean CER

```python
df_cer_clean = df_cer.copy()

df_cer_clean.columns = df_cer_clean.columns.str.strip()

# 统一字段
cer_keep_cols = [
    "project_name",
    "state",
    "capacity_mw",
    "fuel_source",
    "accreditation_code",
    "postcode",
    "accreditation_start_date",
    "approval_date",
    "committed_date",
    "project_status"
]

existing_cer_cols = [c for c in cer_keep_cols if c in df_cer_clean.columns]
df_cer_clean = df_cer_clean[existing_cer_cols].copy()

# 文本清洗
cer_text_cols = [
    c for c in ["project_name", "state", "fuel_source", "accreditation_code", "project_status"]
    if c in df_cer_clean.columns
]

for col in cer_text_cols:
    df_cer_clean[col] = df_cer_clean[col].astype(str).str.strip()

df_cer_clean.replace({"-": pd.NA, "nan": pd.NA, "None": pd.NA}, inplace=True)

# 数值清洗
if "capacity_mw" in df_cer_clean.columns:
    df_cer_clean["capacity_mw"] = pd.to_numeric(df_cer_clean["capacity_mw"], errors="coerce")

# postcode 统一成字符串
if "postcode" in df_cer_clean.columns:
    df_cer_clean["postcode"] = (
        df_cer_clean["postcode"]
        .astype(str)
        .str.replace(".0", "", regex=False)
        .str.strip()
    )
    df_cer_clean.loc[df_cer_clean["postcode"].isin(["", "nan", "<NA>"]), "postcode"] = pd.NA

# 日期清洗
# accredited 相关日期：一般是 日/月/年
if "accreditation_start_date" in df_cer_clean.columns:
    df_cer_clean["accreditation_start_date"] = pd.to_datetime(
        df_cer_clean["accreditation_start_date"],
        format="%d/%m/%Y",
        errors="coerce"
    )

if "approval_date" in df_cer_clean.columns:
    df_cer_clean["approval_date"] = pd.to_datetime(
        df_cer_clean["approval_date"],
        format="%d/%m/%Y",
        errors="coerce"
    )

# committed_date 可能格式不统一，先保守处理
if "committed_date" in df_cer_clean.columns:
    df_cer_clean["committed_date"] = pd.to_datetime(
        df_cer_clean["committed_date"],
        format="%b-%Y",
        errors="coerce"
    )

# 州统一
df_cer_clean["state"] = df_cer_clean["state"].astype(str).str.upper().str.strip()
df_cer_clean.loc[~df_cer_clean["state"].isin(valid_states), "state"] = pd.NA
df_cer_clean = df_cer_clean.dropna(subset=["state"]).copy()

# 去重
df_cer_clean = df_cer_clean.drop_duplicates()

print("=== df_cer_clean info ===")
print(df_cer_clean.info())
print(df_cer_clean["project_status"].value_counts())
df_cer_clean.head()
```

Clean AbS

```python
df_abs_clean = df_abs.copy()
df_abs_clean.columns = df_abs_clean.columns.str.strip()

abs_rename_map = {
    "DATAFLOW": "dataflow",
    "MEASURE: Data Item": "measure",
    "REGIONTYPE: Region Type": "region_type",
    "ASGS_2021: Region": "region",
    "FREQUENCY: Frequency": "frequency",
    "TIME_PERIOD: Time Period": "year",
    "OBS_VALUE": "obs_value",
    "UNIT_MEASURE: Unit of Measure": "unit_measure",
    "UNIT_MULT: Unit of Multiplier": "unit_mult",
    "OBS_STATUS: Observation Status": "obs_status",
    "OBS_COMMENT: Observation Comment": "obs_comment"
}

df_abs_clean = df_abs_clean.rename(columns=abs_rename_map).copy()

df_abs_clean["year"] = pd.to_numeric(df_abs_clean["year"], errors="coerce")
df_abs_clean["obs_value"] = pd.to_numeric(df_abs_clean["obs_value"], errors="coerce")

df_abs_clean = df_abs_clean[
    df_abs_clean["region_type"].astype(str).str.contains("States and Territories", na=False)
].copy()

df_abs_clean["state_name_full"] = (
    df_abs_clean["region"].astype(str).str.split(":", n=1).str[-1].str.strip()
)

state_name_to_code = {
    "New South Wales": "NSW",
    "Victoria": "VIC",
    "Queensland": "QLD",
    "South Australia": "SA",
    "Western Australia": "WA",
    "Tasmania": "TAS",
    "Northern Territory": "NT",
    "Australian Capital Territory": "ACT"
}

df_abs_clean["state"] = df_abs_clean["state_name_full"].map(state_name_to_code)
df_abs_clean = df_abs_clean.dropna(subset=["state"]).copy()

abs_keywords = [
    "Estimated resident population",
    "Land area"
]

mask = df_abs_clean["measure"].astype(str).str.contains("|".join(abs_keywords), case=False, na=False)
df_abs_clean = df_abs_clean[mask].copy()

print("=== df_abs_clean info ===")
print(df_abs_clean.info())
df_abs_clean.head()
```

### State-level summaries

```python
df_nger_state = (
    df_nger_clean.groupby("state", as_index=False)
    .agg(
        nger_record_count=("facility_name", "count"),
        total_generation_mwh=("electricity_production_mwh", "sum"),
        total_emissions_tco2e=("total_emissions_tco2e", "sum"),
        avg_emission_intensity=("emission_intensity_tco2e_per_mwh", "mean")
    )
)

df_cer_state = (
    df_cer_clean.groupby("state", as_index=False)
    .agg(
        cer_project_count=("project_name", "count"),
        total_capacity_mw=("capacity_mw", "sum")
    )
)

df_abs_state = (
    df_abs_clean.groupby("state", as_index=False)
    .agg(
        abs_record_count=("measure", "count"),
        avg_obs_value=("obs_value", "mean")
    )
)

print(df_nger_state)
print(df_cer_state)
print(df_abs_state)
```

### Join by state

```python
df_state_joined = (
    df_nger_state
    .merge(df_cer_state, on="state", how="outer")
    .merge(df_abs_state, on="state", how="outer")
    .sort_values("state")
    .reset_index(drop=True)
)

print("=== Joined state-level table ===")
print(df_state_joined)
```

```python
import pandas as pd
import re

STOP_TOKENS = {
     "farm", "wind", "solar", "stage", "creek", "hill",
    "river", "lake", "park", "energy", "power", "project",
    "facility", "renewable", "generation", "station",
    "green", "hub", "the", "of"
}

def clean_name(x):
    if pd.isna(x):
        return ""
    x = str(x).lower().strip()
    x = re.sub(r"\b(nsw|qld|vic|wa|sa|tas|act|nt)\b", " ", x)
    x = re.sub(r"\b(pty ltd|pty|ltd)\b", " ", x)
    x = re.sub(r"[^a-z0-9]", " ", x)
    x = re.sub(r"\s+", " ", x).strip()
    return x

def token_set(x):
    if not x:
        return set()
    return {t for t in x.split() if t not in STOP_TOKENS and not t.isdigit()}

df_nger["name_key"] = df_nger["Facility name"].apply(clean_name)
df_cer["name_key"] = df_cer["project_name"].apply(clean_name)

matches = []

for _, n_row in df_nger.iterrows():
    n_name = n_row["name_key"]
    n_state = n_row["State"]
    n_tokens = token_set(n_name)

    if not n_tokens:
        continue

    for _, c_row in df_cer.iterrows():
        c_name = c_row["name_key"]
        c_state = c_row["state"]
        c_tokens = token_set(c_name)

        if n_state != c_state:
            continue
        if not c_tokens:
            continue

        overlap = n_tokens & c_tokens

        if len(overlap) >= 1:
            matches.append({
                "nger_name": n_row["Facility name"],
                "cer_name": c_row["project_name"],
                "nger_key": n_name,
                "cer_key": c_name,
                "state": n_state,
                "shared_tokens": ", ".join(sorted(overlap)),
                "overlap_count": len(overlap)
            })

df_match = pd.DataFrame(matches)

print("=== 匹配结果示例 ===")
display(df_match.sort_values("overlap_count", ascending=False).head(20))

print("\n=== 匹配统计 ===")
print("NGER 总数:", len(df_nger))
print("CER 总数:", len(df_cer))
print("匹配对数:", len(df_match))

if len(df_match) > 0:
    print("匹配到的 NGER 数:", df_match["nger_name"].nunique())
    print("匹配到的 CER 数:", df_match["cer_name"].nunique())
```
